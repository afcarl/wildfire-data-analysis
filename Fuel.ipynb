```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Get the fuel and vegetation labels
veg_labels = pd.read_excel("fuel_veg_labels.xlsx", sheetname=0)
fuel_labels = pd.read_excel("fuel_veg_labels.xlsx", sheetname=1) 
```

```python
## Get Fuel data 
escondido_fuel_2010=pd.read_table('Clip_Fuel2010_Escondido.xyz',names=['lat','long','id']
                                   ,delimiter=' ',header=None)
escondido_fuel_2010 = pd.merge(escondido_fuel_2010, fuel_labels, on='id', how='left')


sns.set_style("whitegrid")
ax = sns.barplot(y="label", x="id", data=escondido_fuel_2010)
ax.set_xlabel("Count")
ax.set_ylabel("Fuels")
ax.set_title("Fuel Distribution 2010")
```

```python
## Get Vegetation data
escondido_veg_2010 = pd.read_table('Clip_Veg2010_Escondido.xyz',names=['lat','long','id']
                                   ,delimiter=' ',header=None)
escondido_veg_2010 = pd.merge(escondido_veg_2010,veg_labels, on='id', how='left')
escondido_veg_2010.head()

plt.figure(figsize=(6,8))
sns.set_style("whitegrid")
ax = sns.barplot(y="label", x="id", data=escondido_veg_2010)
ax.set_xlabel("Count")
ax.set_ylabel("Vegetation")
ax.set_title("Vegetation Distribution")
```
