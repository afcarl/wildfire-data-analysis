```python
from scipy.cluster.vq import *
import matplotlib.pyplot as plt
from StringIO import StringIO
from PIL import Image
import numpy as np
from pyspark.mllib.clustering import KMeans, KMeansModel
```

```python
#CHANGE HERE
path = 'kitty.tif'
k = 8

#Get image as binary files and convert to numpy array of RGB values
#If image is 256x256 pixels, it is converted to an array shape of (256x256x3)
images = sc.binaryFiles(path)
inputRDD = images.values().map(lambda rawdata: np.asarray(Image.open(StringIO(rawdata))))

#Flatten RDD to (3 x number_of_pixels)
flattenedRDD = inputRDD.flatMap(lambda xs: [x[0] for x in xs])

model = KMeans.train(flattenedRDD, k=8, maxIterations=10, initializationMode="random")

resultRDD = flattenedRDD.map(lambda x : model.predict(x))

cluster_values = resultRDD.collect()
```
